## Description: This file is used to view the data in the dataset

```python
# Importing the libraries
from matplotlib import pyplot
from matplotlib.image import imread
```

```python
# define location of dataset
folder = "../Image_Downloader/anteater_images/"
# plot first few images
for i in range(9):
 # define subplot
 pyplot.subplot(330 + 1 + i)
 # define filename
 filename = folder + 'Anteater ' + str(i) + '.jpg'
 # load image pixels
 image = imread(filename)
 # plot raw pixel data
 pyplot.imshow(image)
# show the figure
pyplot.show()
```
